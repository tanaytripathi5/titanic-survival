```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv(r'D:\Projects\titanic_survival\train.csv')
```

```python
train_data.info()
```

```python
train_data.drop(['PassengerId','Name','Parch','Ticket','Fare','Cabin','Embarked'],axis=1,inplace=True)
```

```python
train_data.info()
```

```python
X = train_data.iloc[:,1:].values
y = train_data.iloc[:,0].values
```

```python
X = np.array(X)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:,2:3])
X[:,2:3] = imputer.transform(X[:,2:3])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
y.reshape(891,1)
```

```python
X,y = np.array(X),np.array(y)
```

```python
X = np.asarray(X).astype(np.float32)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(5))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss ='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,validation_split=0.2,batch_size=32,epochs=100,verbose=2)
```
